```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'HomeTeam':['A','B','C','D','E','A','B','C','D','E','A','B','C']
     ,'AwayTeam':['C','A','B','E','D','B','D','E','C','A','E','C','A']
     ,'FTHG':[3,2,5,4,2,1,6,2,3,1,2,2,1]
     ,'FTAG':[0,3,6,4,2,0,7,2,4,0,3,4,1]
     , 'FTR' : ['Home','Away','Away','Draw','Draw','Home','Away','Draw','Away','Home','Away','Away','Draw']
    })
```

```python
df
```

```python
#across all seasons thus far
uniqueTeams = df['HomeTeam'].unique()

#build dict with ALL teams; initialize to ----- for each
teamStreak = {uniqueTeams[i]: {'GS': 0,'GC':0,'Streak':'----'} for i in range(0,len(uniqueTeams))}; 

#setup new column and assign it an initial value
# TGS = Total goals scores, TGC = Total goals conceded
df['HomeStreak'] = 'Not set'
df['AwayStreak'] = 'Not set'
df['HTGS'] = 0
df['HTGC'] = 0
df['ATGS'] = 0
df['ATGC'] = 0

```

```python
teamStreak
```

```python
for index,row in df.iterrows():
    
    #Populate streak data into columns before game outcome is counted
    df.at[index,'HomeStreak'] = teamStreak[row['HomeTeam']]['Streak'];
    df.at[index,'AwayStreak'] = teamStreak[row['AwayTeam']]['Streak'];
    df.at[index,'HTGS'] = teamStreak[row['HomeTeam']]['GS'];
    df.at[index,'ATGS'] = teamStreak[row['AwayTeam']]['GS'];
    df.at[index,'HTGC'] = teamStreak[row['HomeTeam']]['GC'];
    df.at[index,'ATGC'] = teamStreak[row['AwayTeam']]['GC'];
    
    teamStreak[row['HomeTeam']]['GS'] += row['FTHG']
    teamStreak[row['AwayTeam']]['GC'] += row['FTHG']
    teamStreak[row['AwayTeam']]['GS'] += row['FTAG']
    teamStreak[row['HomeTeam']]['GC'] += row['FTAG']
    
    #check who won and who lost
    if (row['FTR']=='Draw'):
        teamStreak[row['HomeTeam']]['Streak'] = 'D' + teamStreak[row['HomeTeam']]['Streak']
        teamStreak[row['AwayTeam']]['Streak'] = 'D' + teamStreak[row['AwayTeam']]['Streak']
    elif (row['FTR']=='Home'):
        teamStreak[row['HomeTeam']]['Streak'] = 'W' + teamStreak[row['HomeTeam']]['Streak']
        teamStreak[row['AwayTeam']]['Streak'] = 'L' + teamStreak[row['AwayTeam']]['Streak']
    elif (row['FTR']=='Away'):
        teamStreak[row['HomeTeam']]['Streak'] = 'L' + teamStreak[row['HomeTeam']]['Streak']
        teamStreak[row['AwayTeam']]['Streak'] = 'W' + teamStreak[row['AwayTeam']]['Streak']

    #snipping off last character since above if block added a result to start
    teamStreak[row['HomeTeam']]['Streak'] = teamStreak[row['HomeTeam']]['Streak'][0:-1]
    teamStreak[row['AwayTeam']]['Streak'] = teamStreak[row['AwayTeam']]['Streak'][0:-1]
```
